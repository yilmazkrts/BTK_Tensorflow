```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
dataFrame = pd.read_excel("CO2 Emissions.xlsx")
```

```python
dataFrame
```

```python
turkey_data = dataFrame[dataFrame['country_name'] == 'Turkey']
china_data = dataFrame[dataFrame['country_name'] == 'China']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(turkey_data['year'], turkey_data['value'])
plt.title('Turkey CO2 Emissions Over Years')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (Million Metric Tons)')
plt.grid(True)
```

```python
plt.plot(china_data["year"], china_data["value"])
```

```python
# 'value' sÃ¼tununu sÄ±ralayÄ±n
china_data_sorted = china_data.sort_values(by='value')

# HistogramÄ± oluÅŸturun
plt.hist(china_data_sorted["value"], bins='auto', color='blue', alpha=0.7)
plt.title('CO2 Emissions Distribution in China')
plt.xlabel('CO2 Emissions (Million Metric Tons)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
