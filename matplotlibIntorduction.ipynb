```python
%matplotlib inline 
```

```python
yasListesi = (10,20,30,30,30,40,50,60,60,70)
kiloListesi = (30,92,85,80,75,88,72,68,65,70)
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi)
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Değişim Grafiği")
#plt.show()
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi, "g--")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Değişim Grafiği")
#plt.show()
```

```python
import matplotlib.pyplot as plt

# 2x2'lik bir alt grafik düzeni oluşturalım
plt.subplot(2, 2, 1)  # Birinci alt grafik
plt.plot([1, 2, 3, 4])
plt.title('Alt Grafik 1')

plt.subplot(2, 2, 2)  # İkinci alt grafik
plt.plot([4, 3, 2, 1])
plt.title('Alt Grafik 2')

plt.subplot(2, 2, 3)  # Üçüncü alt grafik
plt.plot([1, 2, 1, 2])
plt.title('Alt Grafik 3')

plt.subplot(2, 2, 4)  # Dördüncü alt grafik
plt.plot([2, 1, 2, 1])
plt.title('Alt Grafik 4')

# Grafikleri göster
plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
numpyDizisi2
```

```python
benimFigur, benimEksen = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2)
```

```python
import matplotlib.pyplot as plt

# Bir figür ve altındaki 2 eksen oluşturma
fig, axs = plt.subplots(2)

# İlk eksen üzerinde çizgi grafiği oluşturma
axs[0].plot([1, 2, 3, 4], [10, 5, 20, 15] )
axs[0].plot([1, 2, 3, 4], [5, 10, 15, 20])
axs[0].set_title('Üstteki Eksen')

# İkinci eksen üzerinde farklı bir grafiği oluşturma
axs[1].scatter([1, 2, 3, 4], [5, 10, 15, 20])
axs[1].set_title('Altındaki Eksen')

# Yükseklik oranını ayarlama
plt.tight_layout()

# Oluşturulan figürü gösterme
plt.show()
```

```python
import matplotlib.pyplot as plt

# Bir figür ve altındaki 2 eksen oluşturma
fig, axs = plt.subplots(2)

# İlk eksen üzerinde çizgi grafiği oluşturma ve etiket ekleme
axs[0].plot([1, 2, 3, 4], [10, 5, 20, 15], label='Çizgi Grafiği')
axs[0].set_title('Üstteki Eksen')
axs[0].legend()  # Etiketleri göstermek için legend kullanma

# İkinci eksen üzerinde farklı bir grafiği oluşturma ve etiket ekleme
axs[1].scatter([1, 2, 3, 4], [5, 10, 15, 20], label='Nokta Grafiği')
axs[1].set_title('Altındaki Eksen')
axs[1].legend()

# Yükseklik oranını ayarlama
plt.tight_layout()

# Oluşturulan figürü gösterme
plt.show()
```

```python
benimFigur, benimEksen = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2 , color = "#3A95A8") #"color" grafiğin çizgisi için renk seçimi yapar
```
